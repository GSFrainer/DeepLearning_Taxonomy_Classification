# Filtering

### Segment Type

```python
raw = raw.loc[raw.gene == "18S_rRNA"]
raw.shape
```

### Sequence Position and Length

```python
raw = raw.loc[raw.start == 1]
raw.shape
```

```python
raw = raw.loc[raw.end >= 900]
raw.shape
```

# Format Rows

```python
tax_levels = ["domain", "supergroup", "division", "subdivision", "class", "order", "family", "genus", "species"]
```

```python
def format_row(row):

    # Remove invalid tax values
    for tax_level in tax_levels:    
        if str(row[tax_level]).endswith("_X") or str(row[tax_level]).endswith("_XX") or str(row[tax_level]).endswith("_XXX") or str(row[tax_level]).endswith("_sp."):
            row[tax_level] = pd.NA
    
    # Trunc sequence
    row["truncated_sequence"] = row["sequence"][0:900]

    return row


#raw.insert(9, "truncated_sequence", pd.NA)
raw = raw.apply(format_row, axis=1)
raw.shape
```
